```python
import pandas as pd
```

```python
movies = pd.read_csv('C:\Datasets/tmdb_5000_movies.csv')
```

```python
credits = pd.read_csv('C:\Datasets/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies.head()
```

```python
# Genres, id, keywords, title, overview, cast, crew
movies = movies[['movie_id', 'title','genres', 'keywords', 'overview', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
movies.head()
```

```python
import ast
#to convert string into list
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count+=1

        else:
            break

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) #converting overview from string to list
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies = movies[['movie_id','title','tags']]
```

```python
new_movies.head()
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() # to make words with same roots the same eg dance,dancing and danced as dance,dance,dance
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
# now we need to convert this textual data into vectors so that we can find the closer vectors to a given vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
cv.fit_transform(new_movies['tags']).toarray().shape
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0][1216]
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
    
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:6]
```

```python
recommend('Batman Begins')
```
